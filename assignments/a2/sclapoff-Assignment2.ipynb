# Problem 1: (15% for CSC 446, 10% for CSC 546)

Suppose a random variable $X$ has the density function as shown the following figure.


![Figure 1](fig1_assn2.png)

1. Write the density function $f(x)$ for this random variable.

Ans: 

$$ f(x)=   \left\{
\begin{array}{ll}
      0 & x<0 \\
      0.7 & 0 \leq x < 1 \\
      0.2 & 1 \leq x < c \\
      0 & x \geq c \\
\end{array} 
\right.  $$

2. What must be the value of $c$?

Ans: 

$$ \int_{R_x} f(x)=1 \\
\int_0^1 0.7dx + \int_1^c 0.2dx = 1 \\
0.7(1) + 0.2(c-1) = 0.5 + 0.2c = 1 \\
c = 0.5/0.2 \\
c = 2.5$$

3. Write the CDF $F(x)$ of this random variable.

Ans: 

$$F(x) = \int_{-\inf}^x f(t)dt \\
F(x) =   \left\{
\begin{array}{ll}
      0 & x<0 \\
      \int_0^x 0.7dt & 0 \leq x < 1 \\
      0.7+ \int_1^x 0.2dt = 0.7+ 0.2(x-1) & 1 \leq x < 2.5 \\
      1 & x \geq 2.5 \\
\end{array} 
\right.  $$
Thus,
$$F(x) =   \left\{
\begin{array}{ll}
      0 & x<0 \\
      0.7x & 0 \leq x < 1 \\
      0.5 +0.2x & 1 \leq x < 2.5 \\
      1 & x \geq 2.5 \\
\end{array} 
\right.  $$

4. What is $E(X)$?

Ans: 

$$E(X) = \int_{-\inf}^{\inf} xf(x)dx \\
E(X) = 0 + \int_0^1 0.7xdx + \int_1^{2.5} 0.2xdx + 0 = 0.7/2 + 0.2\frac{2.5^2-1}{2} \\
E(X) = 0.875$$


5. What is the variance of $X$?

Ans:

$$V(X) = E(X^2) - E(X)^2 \\
E(X^2) = 0 + \int_0^1 0.7x^2dx + \int_1^{2.5} 0.2x^2dx + 0 = 0.7/3 + 0.2\frac{2.5^3-1}{3} = 1.20833 \\
V(X) = 1.20833 - 0.875^2 = 0.442708333 \\
V(X) = 0.4427$$

# Problem 2 (20% for CSC 446, 10% for CSC 546)

A terminal at an airport has one queue to check-in for the Terminal $1$ passengers. After the
check-in, passengers join along with Terminal $2$ passengers to a Security-Check
as shown in the following figure. 


![Figure 1](fig2_assn2.png)

The arrivals are Poisson arrivals with rates $\lambda_1$ and $\lambda_2$ for Terminal $1$ and $2$ passengers, respectively. The service times are exponentially distributed with service rates $\mu_1, \mu_S$ for terminal $1$ check-in and terminal $2$ security-check, respectively.

The average arrival rates are $10$ customers/hour for Terminal $1$ passengers and $20$ customers/hour for Terminal $2$ passengers. The average service times at the Terminal $1$ check-in is $2$ minutes per customer. The security-check maintains a server utilization of $75\%$ all the time.

Answer the following questions with mathematical analysis:

(1) What is the average service time at Security-Check?

Ans: 

From pooling property of Poisson Proccess, average security check (SC) arrival rate, $\lambda_S = \lambda_1 + \lambda_2 = 10+20=30 customers/hr$
$$\mu_S = \lambda_S / \rho_S = \frac{30 customers/hr}{0.75} = 40 customers/hr \frac{hr}{60min} = 2/3 customers/min$$
Average service time = $1/\mu_S = 1.5 min/customer$



(2) What is the the average time spent by the passengers (including both terminal $1$ passengers and terminal $2$ passengers) at Security-Check?

Ans:

SC can be represented by a M/M/1 model, thus time spent in the system, $\hat{\omega_S},$ is:
$$\hat{\omega}_S = \frac{1}{\hat{\mu}_S - \hat{\lambda}_S} = \frac{1}{40 - 30} = \frac{1}{10}hr/customer\frac{60min}{1hr} \\
\hat{\omega}_S = 6min/customer$$

(3) What is the average total time spent in the system by Terminal $1$ passengers(Note that Terminal $1$ passengers need to pass Terminal $1$ and Terminal $2$)?

Ans: 

Average total time spent in the full system by T1 passengers, $\hat{\omega}_{1total} = \hat{\omega}_1 + \hat{\omega}_S$
$$\hat{\omega}_1 = \frac{1}{\hat{\mu}_1 - \hat{\lambda}_1} = \frac{1}{1/(2min/customer) * 60min/hr - 10 customer/hr} = \frac{1}{20cust/hr} \frac{60min}{hr} = 3min/customer  \\
\hat{\omega}_{1total} = 3 + 6 = 9min/customer$$


(4) What is the average total number of Terminal $1$ passengers in the system (Note that Terminal $1$ passengers need to pass Terminal $1$ and Terminal $2$)?

Ans: 

Using Little's law, 
$$\hat{L}_{1tot} = \hat{\lambda}_{1tot} \hat{\omega}_{1tot} = (10customer/hr)(1hr/60min)(9min/customer)\\
\hat{L}_{1tot}  = 1.5 customers$$



(5) What is the average time spent by Terminal $2$ passengers in the system?

Ans: 

$$\hat{L}_{2tot} = \hat{\lambda}_2 \hat{\omega}_S = (20 customer/hr)(6 min/customer)(1hr/60min) \\
\hat{L}_{2tot} = 2 customers$$
